```python
#读取pkl文件查看前5条
import pandas as pd
data = pd.read_pickle('../../POI_data/NYC/category_to_embedding.pkl')
print(data)    

# 计算两个向量的余弦相似度
from numpy import dot
from numpy.linalg import norm
def cosine_similarity(vec1, vec2):
    return dot(vec1, vec2) / (norm(vec1) * norm(vec2))

sim = cosine_similarity(data['Food & Drink Shop'], data['Restaurant'])
print(sim)
```
